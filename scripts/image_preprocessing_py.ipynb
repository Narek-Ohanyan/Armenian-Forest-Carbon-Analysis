```python
import os
import rasterio
import numpy as np
from PIL import Image

# 1. List of the files
my_files = [
    'Dilijan_North_Summer_Peak.tif', 'Dilijan_North_Autumn_Senescence.tif',
    'Dilijan_North_Spring_Rebirth.tif', 'Dilijan_North_Winter_Dormancy.tif',
    'Syunik_South_Summer_Peak.tif', 'Syunik_South_Autumn_Senescence.tif',
    'Syunik_South_Spring_Rebirth.tif', 'Syunik_South_Winter_Dormancy.tif'
]

# 2. Check if files exist
for f in my_files:
    if not os.path.exists(f):
        print(f"❌ ERROR: File '{f}' not found in the sidebar. Please upload it.")
    else:
        print(f"✅ Found: {f}")

# 3. Process the files found
def convert_for_imagej_online(input_tif, output_name):
    try:
        with rasterio.open(input_tif) as src:
            # Read bands (Assuming Band 1 is Red, Band 2 is NIR)
            red = src.read(1).astype('float32')
            nir = src.read(2).astype('float32')

            # Simple NDVI Calculation
            ndvi = (nir - red) / (nir + red + 1e-10)

            # Scale to 8-bit (0-255) for ImageJ Online compatibility
            # This centers the -1 to 1 range so ImageJ can read it easily
            ndvi_8bit = ((ndvi + 1) / 2 * 255).astype(np.uint8)

            im = Image.fromarray(ndvi_8bit)
            im.save(output_name)
            print(f"✨ Created IJ-Ready file: {output_name}")
    except Exception as e:
        print(f"⚠️ Could not process {input_tif}: {e}")

# Run the conversion
for f in my_files:
    if os.path.exists(f):
        convert_for_imagej_online(f, "IJ_" + f)
```
